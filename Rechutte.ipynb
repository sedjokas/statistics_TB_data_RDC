In this part we will average the cases per year and show the year that has had more cases of relapse.

Note: provinces are not considered here

```python
#Declaration and initialization of variables

moyenne=0
annee=0
c=0
tabMoy = []
tabAnn = [] 

#Method to display the years
def monAnne(year):
    an = year[2:len(year)]
    if int(an)<10:
        an="200" + an
        
    else:
        an="20"+an
    return an

#Loop allowing to browse the columns and calculate the average for each year.
for i in dataRc.columns:
    if i=='ZS':             #if the column is that of Health zone or province
        moyenne=moyenne
    else:
        b=dataRc[i].mean()
        tabAnn.append(monAnne(i))
        tabMoy.append(b)
        if b>moyenne:
            moyenne=b
            annee=i
    c+=1

#display   
print("Year ", monAnne(annee), " had the highest average of relapse individual: ",moyenne)
```

```python
#joint table of the year and its case average
tab=[tabAnn,tabMoy]

#transformation of this table into dataframe plus transpose
df=pd.DataFrame(tab).T
df.shape

#nomenclature of columns
df=df.rename(columns={0:'Year',1:'Average'})
print(df)
```

```python
#bar graph of years based on averages
c=df.plot(x=0, y=1, kind='bar')
plt.xlabel('Year')
plt.ylabel('Avarage')
plt.show()

fig = c.get_figure()
fig.tight_layout()
fig.savefig("PhotoMoyRC1.png")
```

In this section we find which province has had many cases for each year.

```python
#Declaration and initialization of variables

maximum=0
annee=0
tabMoy = []
tabAnn = [] 
MaxProv = dict()

#Method to display the years
def monAnne(year):
    an = year[2:len(year)]
    if int(an)<10:
        an="200" + an
        
    else:
        an="20"+an
    return an

for i in dataRc.columns:
    if i=='ZS':
        maximum=maximum
    else:
        b=dataRc[i].max()
        #print(b)
        provIndex=list(dataRc[i]).index(b)
        #print(list(dataNc[i]))
        
        province=dataRc.ZS[provIndex]
        #print(provIndex)
        
        MaxProv[monAnne(i)]=[b,province]
        
#display
for key in MaxProv.keys():
    print('Year: ',key, ' Max: ', MaxProv[key][0], ' Province: ' , MaxProv[key][1])
    print(" ")
```
